OpenCV image tranformation

cv2.resize(image,dsize,fx,fy,interpolation): image size adjust





1.   dsize: manual size
2.   fx : width rate
3.   fy : height rate
4.   interpolation

INTER_CUBIC: Make size bigger
INTER_AREA: Make size smaller

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('bella.jpg')
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.show()

expand = cv2.resize(image,None,fx=2.0,fy=2.0, interpolation=cv2.INTER_CUBIC)
plt.imshow(cv2.cvtColor(expand,cv2.COLOR_BGR2RGB))
plt.show()

shrink = cv2.resize(image,None,fx=0.8,fy=0.8, interpolation=cv2.INTER_AREA)
plt.imshow(cv2.cvtColor(shrink,cv2.COLOR_BGR2RGB))
plt.show()
```

cv2.warpAffine(image,M,dsize): change the location the the image


1.   M: transformation matrix
2.   dsize: manual size



```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('bella.jpg')

height, width = image.shape[:2]

M = np.float32([[1,0,150],[0,1,200]])
dst = cv2.warpAffine(image,M,(width,height))

plt.imshow(cv2.cvtColor(dst,cv2.COLOR_BGR2RGB))
plt.show()
```

cv2.getRotationMatrix2D(center,angle,scale):
transformed matrix to rotate the image


1.   center: center for rotation
2.   angle: angle for rotation
3.   scale: scale factor



```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = cv2.imread('bella.jpg')

height,width = image.shape[:2]

M = cv2.getRotationMatrix2D((width/2,height/2),90,0.5)
dst = cv2.warpAffine(image,M,(width,height))

plt.imshow(cv2.cvtColor(dst,cv2.COLOR_BGR2RGB))
plt.show()
```
