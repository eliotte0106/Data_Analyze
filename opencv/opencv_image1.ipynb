Using OpenCV to check the size and pixel of the image

```python
import cv2
image = cv2.imread('bella.jpg')

#pixel nums and image size
print(image.shape)
print(image.size)

#image numpy pixel
px = image[100,100] #x axis, y axis

#B,G,R
#if gray scale, not B,G,R
print(px)
#R
print(px[2])
```

Using OpenCV to change pixel range

```python
import cv2
import matplotlib.pyplot as plt
import time

image = cv2.imread('bella.jpg')

#method 1
start_time = time.time()
for i in range(0,100):
  for j in range(0,100):
    image[i,j] = [255,255,255]
print("%s seconds" %(time.time()-start_time))
#method 2 : faster
start_time = time.time()
image[0:100,0:100] = [0,0,0]
print("%s seconds" %(time.time()-start_time))

plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.show()
```

OpenCV ROI(Region of interest)

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('bella.jpg')

#numpy slicing: ROI
roi = image[0:400, 300:600]
#ROI image copy
image[0:400, 0:300] = roi

plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.show()
```

OpenCV Pixel color set

```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread('bella.jpg')
image[:,:,2] = 0 # R = 0 (remove R(red))

plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
plt.show()
```
